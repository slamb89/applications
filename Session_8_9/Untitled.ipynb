```python
kean_2 = {}

for emails in os.walk(r"/home/EDSE-sherilb-workspace/Enron_dataset/maildir/kean-s/all_documents/"):
    for mail in emails[2]:
        
        file_to_read = r"/home/EDSE-sherilb-workspace/Enron_dataset/maildir/kean-s/all_documents/" + mail

        with open(file_to_read, "r") as f:
            data = f.read()
    
            email = Parser().parsestr(data)
        
            if email["to"] != None:
        
                email_to = email["to"].strip("\n")

                email_list = email_to.split(",") 

                for m in email_list:

                    if m in kean_2.keys():

                        kean_2[m] += 1

                    else:

                        kean_2[m] = 1
            
        
print(kean_2)
        
        #(type(emails[2]))
   
        #print(emails[2])
```
