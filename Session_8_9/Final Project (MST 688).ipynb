```python
# Created a variable titled "rootdir" which points to the Enron e-mail dataset saved in my EDSE workspace
rootdir = r"/home/EDSE-sherilb-workspace/Enron_dataset/maildir/"

# Defined variable (count_emails) which will list all e-mails in the directory, starting with the first e-mail (0)
count_emails = 0

# Create dictionary

email_sender = {}

# Create empty lists
email_time_sent = []

bad_files = []

bin_0600 = datetime.strptime("06:01:00", "%H:%M:%S").time()

bin_1200 = datetime.strptime("12:01:00", "%H:%M:%S").time()

bin_1800 = datetime.strptime("18:01:00", "%H:%M:%S").time()

bin_0000 = datetime.strptime("23:59:59", "%H:%M:%S").time()

bin_0001 = datetime.strptime("00:00:01", "%H:%M:%S").time()

# Iterating through directory, subdirectory, filenames in rootdir to list the variable "count_emails" which will 
    #also include file names (len(filenames))
for directory, subdirectory, filenames in os.walk(rootdir):
    
    if len(subdirectory) == 150:
 
        #print(len(subdirectory))
    
        for sub in subdirectory:

            #print(sub)

            email_sender[sub] = {"bin_1":0, "bin_2":0, "bin_3":0, "bin_4":0, "bin_5":0}
    

        # Building a pathway to each of the individual e-mails in maildir
    for file in filenames:
        file_to_read = directory + r"/" + file
        
        direct = directory.split(r"/")
        
        name = direct[5]
        

        #print (file_to_read)

        try: 

            with open(file_to_read, "r") as f:
                data = f.read()
        except:

            bad_files.append(file_to_read)
     
        
        email_dates = email_sender[name]

        #Use the parser library to make life easier
        email = Parser().parsestr(data)
        date = email["date"]
        day_of_week = date[0:3]
        #print(date[-20:-12])
        time = datetime.strptime(date[-20:-12],"%H:%M:%S").time()

        if time > bin_0600 and time <= bin_1200:
            email_dates["bin_1"]+=1

        elif time > bin_1200 and time <= bin_1800:
            email_dates["bin_2"]+=1

        elif time > bin_1800 and time <= bin_0000:
            email_dates["bin_3"]+=1

        elif time > bin_0000 and time <= bin_0600:
            email_dates["bin_4"]+=1

        elif time > bin_0001 and time <= bin_0600:
            email_dates["bin_5"]+=1
                

#Use print function to list e-mails
print (email_sender)
```
